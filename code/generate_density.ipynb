```python
import numpy as np

import utils as ut
```

Generate N positions with a uniform radial density. We use the Sample class in the utils.py module

```python
r_min, r_max = 0, 10  # domain
def radial_profile(r, r_min=r_min, r_max=r_max):
    return np.where((r > r_min) & (r < r_max), 1, 0)
sample = ut.Sample(radial_profile, r_max=r_max)

N = 1e5
positions = sample.sample_points(N)
```

Compute the density with a given resolution

```python
sample.compute_density(resolution=100, ranges=[r_max, r_max, r_max])
sample.plot_density()
```

Save it for later reading in blender and converting to .vdb file

```python
sample.save_density('../data/density', float_precision='float64')
```
